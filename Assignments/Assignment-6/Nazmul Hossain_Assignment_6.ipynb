###  Write a function wordcount() that takes the name of a text file as input and prints the number of occurrencesof every word in the file. You function should be case-insensitive so &#39;Hello&#39; and &#39;hello&#39; are treated as the same word. You should ignore words of length 2 or less. Test your implementation on file great_expectations.txt



```python
"""
An example of how it should work would look like this

wordcount(“great_expectations.txt”)
artifice            1
resting             2
compact             1
service             3
sixth               1
lulling             1
radiance            2
offices             1
tree                5
incline             1
treasures           3
unravel             1
oblivion            2
god                 20
malignity           7
promontory          4
acknowledged        1
refinement          1
...
...
"""
 
```

```python
"Complete the code using the template below"
def wordcount(myfile):
    file = open(myfile)
    content = file.read()
    content = content.lower()
     
    import string
    a = list(string.printable)
    for i in string.ascii_lowercase:
        a.remove(i)
    for i in a:
        content = content.replace(i, " ")
    
    content = content.split()
    content = list(content)
    
    new_content = []
    for i in content:
        if (i not in new_content):
            new_content.append(i)
    new_content.sort()
    for i in new_content:
        c = content.count(i)
        print(i, '  ', c)
    
    
wordcount("great_expectations.txt")
```
